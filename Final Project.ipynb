```python
# Step 1: Group by country to calculate the total energy production and percentage of energy production for each country
total_energy_by_country = df.groupby('country_long')['estimated_generation_gwh_2017'].sum().reset_index()
total_energy_by_country['Percentage'] = (total_energy_by_country['estimated_generation_gwh_2017'] / total_energy_by_country['estimated_generation_gwh_2017'].sum()) * 100
total_energy_by_country = total_energy_by_country.sort_values(by='estimated_generation_gwh_2017', ascending=False)
```

```python
print("Total Energy Production by Country:")
total_energy_by_country.head(10)
```

```python
# Step 4: Group by country and fuel type to calculate the percentage of energy production and its contribution to total pollution
energy_pollution_by_country_fuel = df.groupby(['country_long', 'primary_fuel']).agg({'estimated_generation_gwh_2017': 'sum', 'Total MtCO2': 'sum'}).reset_index()
energy_pollution_by_country_fuel['Percentage_Energy'] = (energy_pollution_by_country_fuel['estimated_generation_gwh_2017'] / energy_pollution_by_country_fuel['estimated_generation_gwh_2017'].sum()) * 100
energy_pollution_by_country_fuel['Percentage_Pollution'] = (energy_pollution_by_country_fuel['Total MtCO2'] / df['Total MtCO2'].sum()) * 100
energy_pollution_by_country_fuel = energy_pollution_by_country_fuel.sort_values(by='Total MtCO2', ascending=False)
print("Energy Production and Pollution Contribution by Country and Fuel Type:")
energy_pollution_by_country_fuel.head(50)
```

```python
total_co2_by_country = df.groupby('country_long')['Total MtCO2'].sum().reset_index()
total_co2_by_country['Percentage'] = (total_co2_by_country['Total MtCO2'] / total_co2_by_country['Total MtCO2'].sum()) * 100
total_co2_by_country = total_co2_by_country.sort_values(by='Total MtCO2', ascending=False)

# Store the results in a DataFrame
total_co2_by_country_df = pd.DataFrame(total_co2_by_country)

# Displaying the results
print("Total CO2 Emissions by Country (in Percentage):")
total_co2_by_country_df.head(10)
```

```python
# Assuming df is your DataFrame

# Choose the country of interest
country_name = 'China'

# Filter the DataFrame by the country of interest
country_df = df[df['country_long'] == country_name]

# Group by fuel type to calculate the total energy production for each fuel type
energy_by_fuel_type = country_df.groupby('primary_fuel')['estimated_generation_gwh_2017'].sum().reset_index()
energy_by_fuel_type = energy_by_fuel_type.sort_values(by='estimated_generation_gwh_2017', ascending=False)

# Displaying the results
print(f"Fuel Types and Total Energy Production for {country_name}:")
print(energy_by_fuel_type.head())
```

```python
# Assuming df is your DataFrame

# Select the top 5 countries with the highest total energy production
top_5_countries = total_energy_by_country.nlargest(5, 'estimated_generation_gwh_2017')

# Iterate over each of the top 5 countries
for country_name in top_5_countries['country_long']:
    # Filter the DataFrame by the country of interest
    country_df = df[df['country_long'] == country_name]

    # Calculate the total energy production for the country
    total_energy_country = country_df['estimated_generation_gwh_2017'].sum()

    # Group by fuel type to calculate the total energy production for each fuel type
    energy_by_fuel_type = country_df.groupby('primary_fuel')['estimated_generation_gwh_2017'].sum().reset_index()
    energy_by_fuel_type['Percentage'] = (energy_by_fuel_type['estimated_generation_gwh_2017'] / total_energy_country) * 100
    energy_by_fuel_type = energy_by_fuel_type.sort_values(by='Percentage', ascending=False)

    # Displaying the results for each country
    print(f"Fuel Types and Total Energy Production Percentage for {country_name}:")
    print(energy_by_fuel_type.head())
    print("\n")
```

```python
# Assuming df is your DataFrame

# Calculate the total energy produced by summing up the values from 'estimated_generation_gwh_2017'
total_energy_produced = df['estimated_generation_gwh_2017'].sum()

# Calculate the total CO2 emissions by summing up the values from 'Total MtCO2'
total_co2_emissions = df['Total MtCO2'].sum()

# Print the results
print("Total Energy Produced:", total_energy_produced)
print("Total CO2 Emissions:", total_co2_emissions)
```

```python
with pd.ExcelWriter('Co2_Emissions+Europe_with_groups.xlsx', engine='xlsxwriter') as writer:
    # Write each grouped dataframe to a different sheet
    total_energy_by_country.to_excel(writer, sheet_name='Total Energy by Country', index=False)
    energy_pollution_by_country_fuel.to_excel(writer, sheet_name='Energy Pollution Country Fuel', index=False)
    total_co2_by_country_df.to_excel(writer, sheet_name='Total CO2 by Country', index=False)
    
    # Write the original CSV data to a separate sheet
    df.to_excel(writer, sheet_name='Original Data', index=False)
```
