```python
# Given data
emissions_rate = 1214.5  # g CO2 eq/kWh
energy_production_mwh = 200  # MWh

# Convert emissions rate to g CO2 eq/MWh
emissions_rate_mwh = emissions_rate * 1000  # Convert g CO2 eq/kWh to g CO2 eq/MWh

# Calculate total CO2 emissions in grams
total_CO2_g = emissions_rate_mwh * energy_production_mwh

# Convert total CO2 emissions from grams to metric tons
total_CO2_t = total_CO2_g / 1e6  # Convert grams to metric tons

# Display the result
print("Total CO2 emissions for producing 200 MWh of electricity:", total_CO2_t, "metric tons")
```

```python
# Given data
emissions_rate = merged_df['CO2e/KWh']  # g CO2 eq/kWh
energy_production_gwh = 1  # GWh

# Convert energy production from GWh to MWh
energy_production_mwh = energy_production_gwh * 1000  # Convert GWh to MWh

# Convert emissions rate to g CO2 eq/MWh
emissions_rate_mwh = emissions_rate * 1000  # Convert g CO2 eq/kWh to g CO2 eq/MWh

# Calculate total CO2 emissions in grams
total_CO2_g = emissions_rate_mwh * energy_production_mwh

# Convert total CO2 emissions from grams to metric tons
total_CO2_t = total_CO2_g / 1e6  # Convert grams to metric tons

# Display the result
print("Total CO2 emissions for producing 200 GWh of electricity:", total_CO2_t, "metric tons")
```

```python
# Conversion factor from grams per kilowatt-hour to metric tons per gigawatt-hour
conversion_factor = 1 / (10**6)

# Apply the conversion to the 'CO2e/KWh' column
df['CO2e/GWh'] = merged_df['CO2e/KWh'] * conversion_factor
```
