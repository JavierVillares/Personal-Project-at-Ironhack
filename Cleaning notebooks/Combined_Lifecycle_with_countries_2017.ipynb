```python
url = ('/Users/javi/Desktop/Ironhack/Entregas de github/Personal-Project-at-Ironhack/COUNTRIES_CLEANED.csv')
url2= ('/Users/javi/Desktop/Ironhack/Entregas de github/Personal-Project-at-Ironhack/Supercleaned_Co2_GWH.csv')

df1 = pd.read_csv(url)
df2 = pd.read_csv(url2)
```

```python
df = pd.merge(df1, df2, on='primary_fuel', how='left')

# Calculate CO2 emissions in metric tons
df['Total MtCO2'] = (df['estimated_generation_gwh_2017'] * 1e6) * (df['CO2e/KWh'] / 1e9) * 1000
df = df.round(2)
df.head(19)
```

```python
df.to_csv('2017Co2_Emissions.csv', index=False)
```

```python
# List of European countries
european_countries = ['Albania', 'Andorra', 'Austria', 'Belarus', 'Belgium', 'Bosnia and Herzegovina', 
                      'Bulgaria', 'Croatia', 'Cyprus', 'Czech Republic', 'Denmark', 'Estonia', 'Finland', 
                      'France', 'Germany', 'Greece', 'Hungary', 'Iceland', 'Ireland', 'Italy', 'Kosovo', 
                      'Latvia', 'Liechtenstein', 'Lithuania', 'Luxembourg', 'Malta', 'Moldova', 'Monaco', 
                      'Montenegro', 'Netherlands', 'North Macedonia', 'Norway', 'Poland', 'Portugal', 
                      'Romania', 'Russia', 'San Marino', 'Serbia', 'Slovakia', 'Slovenia', 'Spain', 
                      'Sweden', 'Switzerland', 'Ukraine', 'United Kingdom', 'Vatican City']

# Create a dictionary for mapping
country_mapping = {country: 'Europe' if country in european_countries else country for country in df['country_long']}

# Map the countries in the dataframe
df['country_long'] = df['country_long'].map(country_mapping)
df['country_long'].unique()
df
```

```python
columns_to_drop = ['country','capacity_mw']
df.drop(columns_to_drop, axis=1, inplace=True)
df.info()
df.to_csv('2017Co2_Emissions+Europe.csv', index=False)
```

```python
# Step 1: Group by country and primary fuel type to sum up the total energy produced for each fuel type in each country
grouped_by_country_fuel = df.groupby(['country_long', 'primary_fuel'])['estimated_generation_gwh_2017'].sum().reset_index()

# Step 2: Calculate the total energy produced by each country
total_energy_by_country = grouped_by_country_fuel.groupby('country_long')['estimated_generation_gwh_2017'].sum().reset_index()

# Step 3: Sort the countries by their total energy production in descending order
total_energy_by_country_sorted = total_energy_by_country.sort_values(by='estimated_generation_gwh_2017', ascending=False)

# Step 4: For each country, calculate the percentage of total energy production for each fuel type
for country in total_energy_by_country_sorted['country_long']:
    country_fuel_data = grouped_by_country_fuel[grouped_by_country_fuel['country_long'] == country]
    country_total_energy = total_energy_by_country_sorted[total_energy_by_country_sorted['country_long'] == country]['estimated_generation_gwh_2017'].values[0]
    country_fuel_data['Percentage'] = (country_fuel_data['estimated_generation_gwh_2017'] / country_total_energy) * 100
    country_fuel_data.sort_values(by='Percentage', ascending=False, inplace=True)
    
    # Step 5: Display the results
    print(f"Country: {country}")
    print(country_fuel_data)
    print("\n")
```

```python
# Group the dataframe by 'country_long' and calculate the sum of 'estimated_generation_gwh_2017'
country_energy = df.groupby('country_long')['estimated_generation_gwh_2017'].sum().reset_index()

# Count the number of unique countries
num_countries = len(country_energy)

# Print the number of unique countries
print("Number of unique countries:", num_countries)

# Print the total energy produced for each country
print("\nTotal energy produced by each country:")
country_energy
```

```python
# Calculate the total energy produced
total_energy = country_energy['estimated_generation_gwh_2017'].sum()

# Sort the dataframe by the total energy produced in descending order
country_energy_sorted = country_energy.sort_values(by='estimated_generation_gwh_2017', ascending=False)

# Calculate the percentage of energy produced by each country compared to the total
country_energy_sorted['percentage_of_total'] = (country_energy_sorted['estimated_generation_gwh_2017'] / total_energy) * 100
```

```python
# Sort the dataframe by the highest percentage of total energy produced
country_energy_sorted_by_percentage = country_energy_sorted.sort_values(by='percentage_of_total', ascending=False)

# Print the sorted dataframe by percentage of total energy produced
print("\nEnergy production by percentage of total (sorted by highest percentage):")
country_energy_sorted_by_percentage.head(60)
```
